```python
#2 2019년 영화순위. 
import requests
from bs4 import BeautifulSoup

res = requests.get("https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=2019%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84")
res.raise_for_status
soup = BeautifulSoup(res.text, "lxml")

#극한직업을 보면, a href 아래에 img class = thumb_img 인 게 있다. 이걸 활용해보자
images = soup.find_all("img", attrs={"class":"thumb_img"})

for idx, image in enumerate(images):
    #print(image["src"])             
    #이 경우 http가 있는 것도 있고 http가 없는 것도 있어서 주석처리하고 http를 다 붙여주는 걸로 해본다.
    image_url = image["src"]
    if image_url.startswith("//"):
        image_url = "https:" + image_url
    
    print(image_url)
    
    #페이지에 접속한 후 정보를 파일로 정보하기 위해서 새로운 변수 생성
    image_res = requests.get(image_url)
    image_res.raise_for_status()     

    #파일쓰기
    with open("movie{}.jpg".format(idx+1), "wb") as f:   #영화 포스터 이미지는 텍스트(글자)가 아니므로 wb
        f.write(image_res.content)                       # 해당 리소스가 갖고 있는 정보를 파일로 바로 쓴다. 즉 contend가 이미지
```

```python
#2-1 2019년 영화순위. 1등부터 5등 까지만 가져오기. => idx가 0,1,2,3,4 까지가 5개이므로 이것만 바꿔주면 된다.
import requests
from bs4 import BeautifulSoup

res = requests.get("https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=2019%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84")
res.raise_for_status
soup = BeautifulSoup(res.text, "lxml")

#극한직업을 보면, a href 아래에 img class = thumb_img 인 게 있다. 이걸 활용해보자
images = soup.find_all("img", attrs={"class":"thumb_img"})

for idx, image in enumerate(images):
    #print(image["src"])             
    #이 경우 http가 있는 것도 있고 http가 없는 것도 있어서 주석처리하고 http를 다 붙여주는 걸로 해본다.
    image_url = image["src"]
    if image_url.startswith("//"):
        image_url = "https:" + image_url
    
    print(image_url)
    
    #페이지에 접속한 후 정보를 파일로 정보하기 위해서 새로운 변수 생성
    image_res = requests.get(image_url)          #이미지 소스 url
    image_res.raise_for_status()                 #이미지 소스를 통한 접속

    #파일쓰기. 이게 가능한 이유는 이 주소 자체의 정보가 사진(jpg)이기 때문
    with open("movie{}.jpg".format(idx+1), "wb") as f:   #영화 포스터 이미지는 텍스트(글자)가 아니므로 wb
        f.write(image_res.content)                       # 해당 리소스가 갖고 있는 정보를 파일로 바로 쓴다. 즉 contend가 이미지
                                                        
    #--------------------------------------여기까지 하면 쓸데 없는 것들도 불러와진다.---------------------------------##

    #상위 5개까지만 다운로드 받겠다.
    if idx >= 4:
        break
```

```python
#3-1 2019년 영화순위~2015년까지 . 
#1등부터 5등 까지만 가져오기. => idx가 0,1,2,3,4 까지가 5개이므로 이것만 바꿔주면 된다.
import requests
from bs4 import BeautifulSoup

for year in range(2015, 2020):  #2015~2019까지. 
    #이래야, 2015, 2016...2019년을 입력하는 url 페이지 생성가능
    url = "https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q={}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84".format(year) #2019부분을 변경
    
    res = requests.get(url)
    res.raise_for_status
    soup = BeautifulSoup(res.text, "lxml")

    #극한직업을 보면, a href 아래에 img class = thumb_img 인 게 있다. 이걸 활용해보자
    images = soup.find_all("img", attrs={"class":"thumb_img"})

    for idx, image in enumerate(images):
        #print(image["src"])             
        #이 경우 http가 있는 것도 있고 http가 없는 것도 있어서 주석처리하고 http를 다 붙여주는 걸로 해본다.
        image_url = image["src"]
        if image_url.startswith("//"):
            image_url = "https:" + image_url
        
        print(image_url)
        
        #페이지에 접속한 후 정보를 파일로 정보하기 위해서 새로운 변수 생성
        image_res = requests.get(image_url)          #이미지 소스 url
        image_res.raise_for_status()                 #이미지 소스를 통한 접속

        #파일쓰기. 이게 가능한 이유는 이 주소 자체의 정보가 사진(jpg)이기 때문
        with open("movie_{}_{}.jpg".format(year, idx+1), "wb") as f:       #영화 포스터 이미지는 텍스트(글자)가 아니므로 wb
                                                                     #{}를 2개 넣은 이유는, 덮어쓰기가 될 경우 방지. 즉, 년을 추가하여 새롭게 분류
            
            f.write(image_res.content)                               #해당 리소스가 갖고 있는 정보를 파일로 바로 쓴다. 즉 contend가 이미지


                                                            
        #--------------------------------------여기까지 하면 쓸데 없는 것들도 불러와진다.---------------------------------##

        #상위 5개까지만 다운로드 받겠다.
        if idx >= 4:
            break
```
