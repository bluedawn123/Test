```python
#1
#동적 페이지 크롤링 => 최초 10개 까지만 보여진다. => 이럴 떄 selenium이 필요하다. 
#user-agent로 한국어 페이지로 접근하기 => header 선언 필요
#영화 제목 찍어보기
#10개밖에 못 가져온다. 
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
browser = webdriver.Chrome()
browser.maximize_window()

#페이지 이동
url = "https://play.google.com/store/movies/top"
browser.get(url)

#한번 원하는 방큼 스크롤 내리기  => 자바스크립트 코드가 필요 ㅠ
#browser.execute_script("window.scrollTo(0, 1080)")  #1980X1080 해상도의 경우
#모니터 해상도 높이가 1080인 위치로 스트롤 내리는 것 의미
 
#화면 가장 아래로 스크롤 내리기  => 이것을 반복해야 한다. 높이가 변하지 않을 때까지
#browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

import time
interval = 2 #2초에 한번씩 스크롤 내리기

# 현재 문서 높이를 가져와서 저장
prev_height = browser.execute_script("return document.body.scrollHeight")

#반복 수행
while True:
    #스크룰을 가장 아래로 내림
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    #페이지 2초마다 대기
    time.sleep(interval) 

    #현재 문서 높이를 가져와서 저장
    current_height = browser.execute_script("return document.body.scrollHeight")
    if current_height == prev_height:     #더 이상 로딩되는 게 없을때 반복문 탈출
        break

    prev_height = current_height          #같지 않다 => 새로운 것 로딩 => 다시 한번 더 수행    

print("스크롤 완료")
```

```python
#2. 
#동적 페이지 크롤링 => 최초 10개 까지만 보여진다. => 이럴 떄 selenium이 필요하다. 
#영화 제목 찍어보기
#10개밖에 못 가져온다. 
#selenimu을 통해서 페이지 소스를 가져오기 때문에, url, headers는 필요가 없다. 

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
browser = webdriver.Chrome()
browser.maximize_window()

#페이지 이동
url = "https://play.google.com/store/movies/top"
browser.get(url)

#한번 원하는 방큼 스크롤 내리기  => 자바스크립트 코드가 필요 ㅠ
#browser.execute_script("window.scrollTo(0, 1080)")  #1980X1080 해상도의 경우
#모니터 해상도 높이가 1080인 위치로 스트롤 내리는 것 의미
 
#화면 가장 아래로 스크롤 내리기  => 이것을 반복해야 한다. 높이가 변하지 않을 때까지
#browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

import time
interval = 2 #2초에 한번씩 스크롤 내리기

# 현재 문서 높이를 가져와서 저장
prev_height = browser.execute_script("return document.body.scrollHeight")

#반복 수행
while True:
    #스크룰을 가장 아래로 내림
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    #페이지 2초마다 대기
    time.sleep(interval) 

    #현재 문서 높이를 가져와서 저장
    current_height = browser.execute_script("return document.body.scrollHeight")
    if current_height == prev_height:     #더 이상 로딩되는 게 없을때 반복문 탈출
        break

    prev_height = current_height          #같지 않다 => 새로운 것 로딩 => 다시 한번 더 수행    

print("스크롤 완료")

#스크래핑 작업 시작
soup = BeautifulSoup(browser.page_source, "lxml")  #★★★★★★★★

#div, class 정보 활용
movies = soup.find_all("div", attrs={"class":"ImZGtf mpg5gc"}) 
print("영화의 총 갯수 ? : ", len(movies))  #=> 10으로 출력된다. => 10개 밖에 출력이 안 되는 이유는 이것이 동적페이지 이기 때문이다. 


for movie in movies:
    title = movie.find("div", attrs={"class":"WsMG1c nnK0zc"}).get_text()
    #글자 정보를 뽑기위해 태그를 보니 위와 같은 태그, 클래스. 거기서 .get_text() 사용해서 정보 추출
    print(title)

#10개밖에 못 가져온다. => 실패. 
#10개까지 밖에 못 가져오는 이유는 태그들이 복잡하게 다른 형식. 상위10와 11개부터 태그들이 다른 형식이다. 
```

```python
#상위 10개와 하위가 태그 등이 다른 경우
#2
#동적 페이지 크롤링 => 최초 10개 까지만 보여진다. => 이럴 떄 selenium이 필요하다. 
#user-agent로 한국어 페이지로 접근하기 => header 선언 필요
#영화 제목 찍어보기
#10개밖에 못 가져온다. 
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
browser = webdriver.Chrome()
browser.maximize_window()

#페이지 이동
url = "https://play.google.com/store/movies/top"
browser.get(url)

#한번 원하는 방큼 스크롤 내리기  => 자바스크립트 코드가 필요 ㅠ
#browser.execute_script("window.scrollTo(0, 1080)")  #1980X1080 해상도의 경우
 
#한번 화면 가장 아래로 스크롤 내리기  => 이것을 반복해야 한다. 높이가 변하지 않을 때까지
#browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

import time
interval = 2 #2초에 한번씩 스크롤 내리기

# 현재 문서 높이를 가져와서 저장
prev_height = browser.execute_script("return document.body.scrollHeight")

#반복 수행
while True:
    #스크룰을 가장 아래로 내림
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")

    time.sleep(interval)

    #현재 문서 높이를 가져와서 저장
    current_height = browser.execute_script("return document.body.scrollHeight")
    if current_height == prev_height:
        break

    prev_height = current_height

print("스크롤 완료")

#스크래핑 작업 시작

soup = BeautifulSoup(browser.page_source, "lxml")


#div, class 정보 활용
movies = soup.find_all("div", attrs={"class":"Vpfmgd"}) #★★ 리스트 활용
#이렇게 하면 클래스가 list 안에 있는 것과 일치하는 것, 즉 둘 다 가져오는 것 의미

'''해설 =>
#상위 10개의 정보를 보면,
movies = soup.find_all("div", attrs={"class":"ImZGtf mpg5gc"}) 이렇게 되어있다.
하지만 하위를 보면, c-wiz 태그 이하에, "div", attrs={"class":"Vpfmgd"로 되어있다.   
'''
                                                                           
print("영화의 총 갯수 ? : ", len(movies))  #=> 10으로 출력된다. => 10개 밖에 출력이 안 되는 이유는 이것이 동적페이지 이기 때문이다. 

for movie in movies:
    title = movie.find("div", attrs={"class":"WsMG1c nnK0zc"}).get_text()
    print(title)
    
#완성.
```
