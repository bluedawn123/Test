```python
#네이버 시가총액 1등 200등 정보를 csv파일로 저장하기
# 
#전과는 다른 방식 => 개발자 도구 확인하니 table 전체를 가져오는 방식. 그 이후 tbody가져오기
#타이틀 넣기
```

```python
import csv
import requests
from bs4 import BeautifulSoup

url = "https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page="
#페이지를 확인해보니, page뒤 숫자(1,2,3...)에 따라 변한다. 전보다 쉬운 케이스

for page in range(1, 5): #1페이지부터 4페이지
    res = requests.get(url + str(page))
    res.raise_for_status()
    soup = BeautifulSoup(res.text, "lxml")

    #데이터 구성 가져오기 전에 형태 먼저 확인해보기  
    #table로 구성.
    #table을 가져오고, table 밑에서 tr을 가져와서 그 밑의 td들을 가져와보자. 
    #thead의 tr은 필요가 없다. tbody의 tr만 필요하다.

    data_rows = soup.find("table", attrs={"class":"type_2"}).find("tbody").find_all("tr")
                          #테이블    #테이블의 class정보       #tbdby가져오기   #tr가져오기
                          #list형태로 저장

    for row in data_rows:
        columns = row.find_all("td")   
        #td정보 가져오기 => 우리가 필요한 정보는 현재가, 시가총액, 종목명 등등 이므로 get_text()를 사용해 가져올 수 있다.
        
        if len(columns) <= 1: #의미없는 데이터 스킵 => 줄바꿈의 경우, 하나의 tr이 하나의 td를 갖는 경우 스킵을 하면 된다.. 의미없는 줄바꿈은 생략
            continue                
        
        data = [column.get_text().strip() for column in columns]  
        #=> columns에 있는 것들을 가져와서 column이라고 하고 그것의 get_text()를 뽑아 td들이 가지고 있는 정보를 저장한다. 
        #strip()를 사용하여 \n\n\n\n등등 제거 가능
        
        #print(data)              #찍어서 보니 불필요한 것들이 많아서 정리를 해야한다. 위의 if len(columns) 참조
        
        print(data)


        ###정리
        #=> 출력해서 보면, '\n\n\t\t\t\t105\n\t\t\t\t\n', '\n\n\t\t\t\t+0.62%\n\t\t\t\t\n', [''][''][''] 등과 같이 줄바꿈, 빈공간이 있기 때문에 이것을                없애야 한다.
        #=> 줄바꿈의 경우, 하나의 tr이 하나의 td를 갖는 경우 스킵을 하면 된다.
        #=> strip() => 문자열 양쪽에 있는 한 칸 이상의 연속된 공백을 모두 지운다.
```
