```python
import requests
from bs4 import BeautifulSoup

url = "https://comic.naver.com/webtoon/weekday.nhn"
res = requests.get(url)
res.raise_for_status()  #혹시 문제시 프로그램 종료


soup = BeautifulSoup(res.text, "lxml")  #html문서값과 lxml parser입력. 
                                        #가져온 html문서를 lxml 파서를 통해서 Beautifulsoup 객체로 만들었다.
                                        #이때 객체인 soup은 모든 정보를 갖고 있다.

print(soup.title)                       #title정보 출력 -> <title>네이버 만화 &gt; 요일별  웹툰 &gt; 전체웹툰</title>
print(soup.title.get_text())            #title.get_text정보 출력 -> 네이버 만화 > 요일별  웹툰 > 전체웹툰

print(soup.a)                           #soup은 모든 html을 가지고 있다. 즉 soup객체에서 발견되는 첫 a에 대한 정보를 뿌려줘라는 의미

#a href="#menu" onclick="document.getElementById('menu').tabIndex=-1;document.getElementById('menu').focus();return false;"><span>메인 메뉴로 바로가기</span></a> ===>>> ★★★★★soup객체의 a에 해당되는 속성 정보값만 출력

print(soup.a.attrs)   #a의 속성 정보 반환(출력)
                      #{'href': '#menu', 'onclick': "document.getElementById('menu').tabIndex=-1;document.getElementById('menu').focus();return                               false;"} => 딕셔너리 형태
                      
                      #soup.a 와 soup.a.attrs의 차이를 알아야 한다. 미세하지만 중요

print("soup.a[href] : ", soup.a["href"])    #soup객체의 a속성 중, href의 속성만 보여준다  #menu
                  

####----------------------------------------특정 a태그의 내용을 찾는 방법----------------------------------#############

print("\n==>웹툰 올리기 정보 가져오기")                   #정보 확인을 해 보면 a태그고 클래스가 Nbtn_upload이다. 


soup.find("a", attrs={"class":"Nbtn_upload"})           #soup 객체 내에서 해당하는 첫 엘리먼트를 가져온다. 이후는 조건의미 attrs=속성, attrs=                                                                      {"class":"Nbtn_upload"}
                                                        #즉, soup 객체 내에서 첫 a태그에 해당하면서 클래스의 속성 정보가 "Nbtn_upload" 인 것을 찾아줘
                                                        
print(soup.find("a", attrs={"class":"Nbtn_upload"}))  #<a class="Nbtn_upload" href="/mypage/myActivity.nhn" onclick="nclk_v2(event,'olk.upload');                                                           ">웹툰 올리기</a>

#soup.find(attrs={"class":"Nbtn_upload"})       =>전체중, class":"Nbtn_upload" 어떤 element를 찾아줘
#soup.find("a", attrs={"class":"Nbtn_upload"})  =>class":"Nbtn_upload" 인 a element를 찾아줘란 의미


####-----------------------------      인기급상승 만화 1등~10등 정보 다양하게 가져오기      ----------------------------

####방법1. li class="rank01" => li태그이고 class가 rank01인 것을 가져오면 된다.
print("\n==>인기 급상승 만화 1등 정보 가져오기 방법1")
print(soup.find("li", attrs={"class":"rank01"}))
'''
<li class="rank01">
<a href="/webtoon/detail.nhn?titleId=641253&amp;no=346" onclick="nclk_v2(event,'rnk*p.cont','641253','1')" title="외모지상주의-346화 일해회 (3계열사) [마무리]">외모지상주의-346화 일해회 (3계열사) [마무리]</a>
<span class="rankBox">
<img alt="변동없음" height="10" src="https://ssl.pstatic.net/static/comic/images/migration/common/arrow_no.gif" title="변동없음" width="7"/> 0							</span>
</li>
이렇게 다 출력된다.'''

####방법2. 변수에 객체 저장 후, 객체.a 활용
print("\n==>인기 급상승 만화 1등 정보 가져오기 방법2")
rank1 = soup.find("li", attrs={"class":"rank01"})  #=> 이게 더 유용하다. soup.find를 통해 발견된 객채를 변수(rank1)에 저장하고
print(rank1.a)                                     #=> rank1.a를 찍어본다. 즉, rank1의, a의 속성 정보 반환(출력)
'''
<a href="/webtoon/detail.nhn?titleId=641253&amp;no=346" onclick="nclk_v2(event,'rnk*p.cont','641253','1')" title="외모지상주의-346화 일해회 (3계열사) [마무리]">외모지상주의-346화 일해회 (3계열사) [마무리]</a>
'''

####방법3.부모, 자식관계에서 가져오기. (rank1에서 rank2를 가져오기)
print("\n==> 부모, 자식관계에서 가져오기. (rank1에서 rank2를 가져오기)")
print("rank1의 a의 글자를 가져오기 : ", rank1.a.get_text())  # => rank1의 a의 글자 부분만 가져오기 : 외모지상주의-346화 일해회 (3계열사) [마무리]

print(rank1.next_sibling)                #태그 사이의 개행정보...즉...줄바꿈 때문에 아무것도 안 나온다. 이럴 경우는 2번 해주면 된다.       =====>>> find_next_sibling() 을 활용하면 된다.
print("rank2의 아래를 sibling으로 가져오기 : ", rank1.next_sibling.next_sibling)   #2위인 바른연애 길잡이에 대한 내용이 아래처럼 잘 나온다. 
                                                                                 #그러므로 이어서 방법2인 변수에 객체 저장 하는 방법을 활용한다.
'''정보 확인.
rank2의 아래를 sibling으로 가져오기 :  <li class="rank02">
<a href="/webtoon/detail.nhn?titleId=772764&amp;no=7" onclick="nclk_v2(event,'rnk*p.cont','772764','2')" title="죽지 않으려면-7화">죽지 않으려면-7화</a>
<span class="rankBox">
<img alt="변동없음" height="10" src="https://ssl.pstatic.net/static/comic/images/migration/common/arrow_no.gif" title="변동없음" width="7"/> 0
'''

rank2 = rank1.next_sibling.next_sibling
rank3 = rank2.next_sibling.next_sibling
print("rank3의 text : ", rank3.get_text())                  #3번 랭크의 글자가 잘 나온다. 한림체육관

#previous를 활용한 전자의 정보 가져오기
rank2 = rank3.previous_sibling.previous_sibling             #한림체육관 이전의 바른연애 길잡이 가져오기
print("previous를 활용한 전자 정보 가져오기 : ", rank2.get_text())

#부모 정보 가져오기
print("rank1의 부모 : ", rank1.parent)   #rank1이 li태그임을 알 수 있다.

#★☆개행 문제 해결 방법★☆
rank2 = rank1.find_next_sibling("li")   #괄호안은 조건. li태그 안에 있는 다음 항목들만 찾아준다.(개행 정보는 찍히지 않는다.))
print("find_next_sibling을 활용한 rank2 정보 가져오기 : ", rank2.a.get_text())    #바른연애 길잡이-141
rank3 = rank2.find_next_sibling("li")
print("find_next_sibling을 활용한 rank3 정보 가져오기 : ", rank3.a.get_text())    #한림체육관-42화
rank2 = rank3.find_previous_sibling("li")
print("find_previous_sibling을 활용한 rank2 정보 가져오기 : ", rank2.a.get_text())    #바른연애 길잡이-141


#☆★★☆★☆---------------------------한번에 모든 정보 가져오기 --------------------------------- => find_next_siblings
print("\n==>한번에 원하는 정보를 가져오기")
#rank1.find_next_siblings
print("find_next_siblings를 활용한 형제 정보들 가져오기 : ", rank1.find_next_siblings("li"))  #rank1의 li의 태그의 모든 정보를 가져온다.

```

```python
#find의 더 많은 용법
import requests
from bs4 import BeautifulSoup

url = "https://comic.naver.com/webtoon/weekday.nhn"
res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml") 

webtoon = soup.find("a", text="바른연애 길잡이-141")  
#전체(soup)에서 a태그를 찾을건데, text라는 값을 비교할 수 있다.
#

#개발자 도구에서 a태그의 정보를 복붙한 것임.
#아래의 a태크 보면, 
#  <a onclick="nclk_v2(event,'rnk*p.cont','703852','2')" 
#  href="/webtoon/detail.nhn?titleId=703852&amp;no=141" 
#  title="바른연애 길잡이-141">바른연애 길잡이-141</a>     #=> 여는 태그와 닫는 태크. 텍스트는 여는 태그와 닫는 태그의 사이의 '바른연애 길잡이-141'이다. 

#다시 말하자면, <a  와   </a> 사이에 있는, 바른연애 길잡이-141가 즉 텍스트 정보이다.


print(webtoon)  
#모든 정보중에서 엘리먼트가 a이고 텍스트가 바른연애 길잡이-141인 것을 찾아란 의미
#<a href="/webtoon/detail.nhn?titleId=703852&amp;no=141" onclick="nclk_v2(event,'rnk*p.cont','703852','2')" title="바른연애 길잡이-141">바른연애 길잡이-141</a>

#이런식으로도 할 수 있다~~
```
