```python
from selenium import webdriver

browser = webdriver.Chrome("./chromedriver.exe")   #chromeDriver경로가 다를 경우 안에 경로를 적어줘야 한다. 같은 경로에 있을 경우는 필요없긴함
browser.get("http://naver.com")                    #chrome.webdriver객체를 생성하고, 그 브라우저에서 url로 이동하는 작업       

#로그인 버튼 누르기  python 실행창에서는 하나씩 바로 입력이 된다. 
element = browser.find_element_by_class_name("link_login")
          #클래스 이름이 개발자도구에서 link_login이었는데, 이 경우
          #browser.find_element_by_class_name으로 검색하면 된다.

element
element.click()
browser.back()  #이전으로 돌아가기. browser.forward()는 앞 페이지로 가기


#검색창 응용
elem = browser.find_element_by_id("query")
       #개발자도구 검색해보면, input id="query". 즉 쿼리라는 아이디를 갖은 input 엘리먼트 검색
       #위와 유사
elem

#키입력을 위해서
from selenium.webdriver.common.keys import Keys  #키입력
elem.send_keys("윤준호")                          #윤준호 치기
elem.send_keys(Keys.ENTER)                       #엔터 치기
elem = browser.find_elements_by_tag_name("a")   
#a태그의 어떤 값을 가져오고 싶을때. a태그가 많을 것이다, 이때 모든 element를 가져오고 싶은 경우 elements로

#elem         #=> 실행시 수많은 정보가 나온다. 

for e in elem: #beautiful soup에서는 href 정보를 가져오려면 elem.get_attribute(["href"])                    
               #하지만, 셀레늄은 대괄호가 필요 없다. 
    
    e.get_attribute("href")   #이렇게 하면,elem의 특성값을 가져올 수 있다. 여기선 href특성(속성)
                              #=> 실행시 수많은 정보가 나온다. 
    
    print(e)   #a태크의 해당하는 element들의 href 속성 정보를 가져온다. 

browser.get("http://daum.net")
elem = browser.find_element_by_name("q")  #검색창. name이 q이다. 
elem.send_keys("윤준호")
#X-paht로 검색버튼 누르기
elem = browser.find_element_by_xpath("//*[@id='daumSearch']/fieldset/div/div/button[2]")
elem.click()
browser.close() #해당 브라우저 닫기
browser.quit()  #모든 브라우저 닫기

#오류가 나도 괜찮음.
```
