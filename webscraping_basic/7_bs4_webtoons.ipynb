```python
#네이버 웹툰 수요일 목록 가져오기
# class 속성이 title인 모든 "a" element를 반환하는 것
import requests
from bs4 import BeautifulSoup

url = "https://comic.naver.com/webtoon/weekday.nhn"  #네이버 웹툰의 url을 가져와서 soup객체를 만들었다.
res = requests.get(url)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml") 

cartoons = soup.find_all("a", attrs={"class":"title"})  #find는 조건의 첫번째 element만 찾는거고 find_all은 모든 element를 찾는 것
                                                        #태그명이 a이고 클래스 속성이 title인 것을 모두 가져온다.

#<a href="/webtoon/list.nhn?titleId=747269&amp;weekday=wed" onclick="nclk_v2(event,'thm*w.tit','','2')" class="title" title="전지적 독자 시점">전지적 독자 시점</a>
print(cartoons)

for cartoon in cartoons:
    print(cartoon.get_text())  #a태그에 있는 get_text()를 통해 글자만 빼오는 것.
```
