```python
import pymysql
import pandas as pd
```

```python
host_name = 'localhost'
host_port = 3306
username = 'root'
password = '123123'
database_name = 'student_mgmt'

#변수로 설정
```

```python
db = pymysql.connect(
    host=host_name,     # MySQL Server Address
    port=host_port,     # MySQL Server Port
    user=username,      # MySQL username
    passwd=password,    # password for MySQL username
    db=database_name,   # Database name
    charset='utf8'
)
```

```python
##### pandas.read_sql(쿼리, 연결된 db connection 객체)
```

```python
SQL = "SHOW TABLES"
```

```python
df = pd.read_sql(SQL, db)
#두 개의 인자가 필요하다. 하나는 쿼리(sql), 하나는 db연결 객체
```

```python
SQL = "SELECT * FROM students"
df = pd.read_sql(SQL, db)
df.to_csv('students.csv', sep=',', index=False, encoding='utf-8')   #to_csv함수 이용해서 csv파일로 내보내는 방법
df
```
